# Built-in data structures

## List
- just like arrays in Java, with more flexibility
- ordered, mutable
- can contain different data types
- more convenient to map each element in the list - List comprehension


````python
random_list = [1, 2, 3]
random_list = [x + 1 for x in random_list if x > 1]
random_list
````

## Dictionary

- one of the most useful data structure in Python
- contains key and value pairs
- like Hashmap in Java or object in javascript, but with higher flexibility

````python
random_dict = { 'a': 1 }
print(random_dict['a']) # 1
random_dict['b'] = 2;
random_dict
````

## Sets

- Similar to Set class in java
- unordered, mutable
- All elements in a set is unique (no duplication)
- Useful when you want to get the unique values or to remove duplicated element


````python
random_set = { 'a', 'a', 'b', 'c' }
print(random_set)
````

## Tuple
- Immutable, ordered
- Used frequently to group elements up
- Small collections

````python
random_tuple = [('g', 'h'), ('5', '6')]
random_tuple
````

## Side-by-side comparison

| List | Tuple | Set |
| ---- | ----- | --- |
| Mutable | Immutable | Mutable |
| Ordered | Ordered | Unordered |


# Functions

- Similar to methods in Java
- Taking an input (parameter), process it and output the processed value (returned value)
- Functions are to group operations, so that it is easy to debug


````python
def addition(x, y):
	z = x + y
	return z
````

````python
addition(1, 2)
````

# Lambda functions

- Similar to lambda functions in Java or anonymous functions in Javascript
- Simple functions without names


```
// Java
Collections.sort(numbers, (n1, n2) -> n1.compareTo(n2));
```

````python
random_list = [1, 3, 5]
add_one = lambda x: x + 1
print([add_one(x) for x in random_list])
````
