
# What is NumPy?
NumPy stands for Numerical Python, which is the fundamental package for scientific computing in Python. It supports for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.


# Why use NumPy?
*   Faster and more powerful than lists
*   Supports basic matrix operations
*   Supports statistics operands
*   Supports broadcasting
*   Supports vectorization

However, NumPy arrays  (as known as ndarrays) comes with some restrictions:

*   NumPy arrays have a fixed size at creation, unlike Python lists (which can grow dynamically). Changing the size of an ndarray will create a new array and delete the original.
*   The elements in a NumPy array are all required to be of the same data type, and thus will be the same size in memory. 




# Get started with Numpy
Let's import the NumPy library and create our first NumPy array.

```python
import numpy as np
np.__version__
```

```python
# create 1-D arrays
my_arr = np.array([1, 2, 3, 4, 5])
print(my_arr)
print(type(my_arr))
print(my_arr.shape)
print(my_arr.size)
print(my_arr.ndim)
```

```python
# create 2-D arrays
my_arr2D = np.array([[1, 2, 3], [4, 5, 6]])
print(my_arr2D)
print(type(my_arr2D))
print(my_arr2D.shape)
print(my_arr2D.size)
print(my_arr2D.ndim)
```

```python
# create 3-D arrays
my_arr3D = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])
print(my_arr3D)
print(type(my_arr3D))
print(my_arr3D.shape)
print(my_arr3D.size)
print(my_arr3D.ndim)
```

# Other functions

**Mathematical functions**
*   sum, prod, cumsum
*   min, max, argmin, argmax
*   ptp (max - min)

**Statistics**
*   mean, std, var

```python
a = np.arange(1,7).reshape(2,3)
a
```

```python
# sum
# by default sum will add up all the values in the array
print(a.sum())
# you can explicitly specific which axis you intend to sum along
print(a.sum(axis=0))
print(a.sum(axis=1))
```

```python
# prod
print(a.prod())
print(a.prod(axis=0))
```

```python
# cumsum
print(a.cumsum())
print(a.cumsum(axis=0))
```

```python
# min
print(a.min())
print(a.min(axis=0))
```

```python
# max
print(a.max())
print(a.max(axis=0))
```

```python
# argmin & argmax
# sometimes we are not interested in the value of min/max, but instead the location of it
# these arg methods return the location in 1D, on a raveled index of arrays
print(a.argmin())
print(a.argmax())
```

```python
# the following code is used to unflatten the 1D location
print(np.unravel_index(a.argmin(), a.shape))
print(np.unravel_index(a.argmax(), a.shape))
```

```python
# ptp
print(a.ptp())
print(a.ptp(axis=1))
```

```python
# statistics functions
print(a.mean())
print(a.std())
print(a.var())
```

# Broadcasting
NumPy arrays of different dimensionality can be combined in the same expression. Arrays with smaller dimension are broadcasted to match larger arrays without copying the data.

```python
a = np.array([0,10,20,30])
b = np.array([4,5,6])
a = a[:, np.newaxis]
print(a)
print(b)
```

```python
a+b
```

Broadcast.jpg

```python
# Note: The trailing axes of either arrays must be 1 or both have the same size for broadcasting to occur
error_a = np.arange(25).reshape(5,5) # 5 x 5 
error_b = np.arange(24).reshape(6,4) # 6 x 4
print(error_a)
print(error_b)
print(error_a+error_b)
```
